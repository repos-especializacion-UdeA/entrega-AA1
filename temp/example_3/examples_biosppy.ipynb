```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros de la señal
frecuencia_seno = 2  # Frecuencia de la señal seno (Hz)
frecuencia_muestreo = 100  # Frecuencia de muestreo (Hz)
duracion = 1  # Duración de la señal en segundos

# Generar el tiempo de la señal de acuerdo a la frecuencia de muestreo
t = np.linspace(0, duracion, int(frecuencia_muestreo * duracion), endpoint=False)

# Generar la señal seno
señal_seno = np.sin(2 * np.pi * frecuencia_seno * t)

# Graficar la señal
plt.figure(figsize=(10, 4))
plt.plot(t, señal_seno, label='Señal Seno 2Hz')
plt.title('Señal Senoidal de 2Hz Muestreada a 100Hz')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()
plt.show()
```
